1) Un empleado trabaja 48 horas a la semana a razón de $5.000 hora. El porcentaje de retención en la fuente es del 12,5% del salario bruto (salario bruto es el total antes de realizar los descuentos y el neto es el salario después de los descuentos). Se desea saber cuál es el salario bruto, la retención en la fuente y el salario neto del trabajador.
 en python

```python
HT = 48
Pago_hora = 5000
porcentaje_retencion = 12.5

salario_bruto = HT * Pago_hora

retencion_fuente = salario_bruto * (porcentaje_retencion / 100)

salario_neto = salario_bruto - retencion_fuente

print("Salario Bruto: $", salario_bruto)
print("Retención en la Fuente: $", retencion_fuente)
print("Salario Neto: $", salario_neto)
```

3) Dado el valor del lado en un triángulo equilátero, haga un algoritmo que obtenga el perímetro, el valor de la altura y el área del triángulo.


```python
import math

# Triángulo equilátero
lado = 25

perimetro = 3 * lado

altura = (math.sqrt(3) / 2) * lado

area = (math.sqrt(3) / 4) * lado ** 2

print("Perímetro del triángulo equilátero:",perimetro)
print("Altura del triángulo equilátero:",altura)
print("Área del triángulo equilátero:",area)
```

4) Un estudiante desea saber cuál será su promedio general y el promedio específico de dos de sus materias. Estas se evalúan como se muestra a continuación:

i. La calificación de Cálculo se obtiene de la siguiente manera:

* Examen 90%
* Promedio de trabajos 10%
* En esta materia se pidieron un total de tres trabajos.

ii. La calificación de Programación se obtiene de la siguiente manera:

* Examen 80%
* Promedio de trabajos 20%
* En esta materia se pidieron un total de dos trabajos.


```python
examen_calculo = float(input("Por favor ingrese la nota del examen de Cálculo: "))
t1_calculo = float(input("Por favor ingrese la nota del primer trabajo: "))
t2_calculo = float(input("Por favor ingrese la nota del segundo trabajo: "))
t3_calculo = float(input("Por favor ingrese la nota del tercer trabajo: "))

examen_programacion = float(input("Por favot ingrese la nota del examen de Programación: "))
t1_programacion = float(input("Por favor ingrese la nota del primer trabajo: "))
t2_programacion = float(input("Por favor ingrese la nota del segundo trabajo: "))

# promedio Cálculo
promedio_trabajos_calculo = (t1_calculo + t2_calculo + t3_calculo) / 3
calificacion_calculo = (examen_calculo * 0.90) + (promedio_trabajos_calculo * 0.10)

# promedio Programación
promedio_trabajos_programacion = (t1_programacion + t2_programacion) / 2
calificacion_programacion = (examen_programacion * 0.80) + (promedio_trabajos_programacion * 0.20)

promedio_general = (calificacion_calculo + calificacion_programacion) / 2

print("Promedio de Cálculo: ",calificacion_calculo)
print("Promedio de Programación: ",calificacion_programacion)
print("Promedio general: ",promedio_general)
```

6) Tres personas deciden invertir su dinero para fundar una empresa. Cada una de ellas invierte una cantidad distinta. Obtener el porcentaje que cada quien invierte con respecto a la cantidad total invertida.

```python
inversion_p1 = float(input("Ingrese la cantidad invertida por la primera persona: "))
inversion_p2 = float(input("Ingrese la cantidad invertida por la segunda persona: "))
inversion_p3 = float(input("Ingrese la cantidad invertida por la tercera persona: "))

cantidad_total_invertida = inversion_p1 + inversion_p2 + inversion_p3

# Calcular el porcentaje invertido por cada persona
porcentaje_p1 = (inversion_p1 / cantidad_total_invertida) * 100
porcentaje_p2 = (inversion_p2 / cantidad_total_invertida) * 100
porcentaje_p3 = (inversion_p3 / cantidad_total_invertida) * 100

# Imprimir los resultados
print("Porcentaje invertido por la primera persona:", porcentaje_p1,"%")
print("Porcentaje invertido por la segunda persona:", porcentaje_p2,"%")
print("Porcentaje invertido por la tercera persona:", porcentaje_p3,"%")
```

9) Elaborar un algoritmo que calcule e imprima:

* El total vendido en la Semana
* El promedio vendido el miércoles
* El promedio vendido en la Semana
* Tomando como base el número de ventas de la semana, calcular y mostrar el porcentaje de ventas cada día


```python
# Ventas por día
NL_ventas = int(input("Por favor ingrese el número de ventas del lunes: "))
Totlunes  = float(input("Por favor ingrese el total vendido del lunes: "))

NMa_ventas = int(input("Por favor ingrese el número de ventas del martes: "))
Totmartes = float(input("Por favor ingrese el total vendido del martes: "))

NMI_ventas = int(input("Por favor ingrese el número de ventas del miércoles: "))
Totmiercoles = float(input("Por favor ingrese el total vendido del miércoles: "))

NS_ventas = int(input("Por favor ingrese el número de ventas del sábado: "))
Totsabado = float(input("Por favor ingrese el total vendido del sábado: "))

total_semana = Totlunes  + Totmartes + Totmiercoles + Totsabado

# promedio vendido el miércoles
if NMI_ventas != 0:
    promedio_miercoles = Totmiercoles / NMI_ventas
else:
    promedio_miercoles = 0

# promedio vendido en la semana
total_ventas = NL_ventas + NMa_ventas + NMI_ventas + NS_ventas
if total_ventas != 0:
    promedio_semana = total_semana / total_ventas
else:
    promedio_semana = 0

# porcentaje de ventas cada día
if total_ventas != 0:
    porcentaje_lunes = (NL_ventas / total_ventas) * 100
    porcentaje_martes = (NMa_ventas / total_ventas) * 100
    porcentaje_miercoles = (NMI_ventas / total_ventas) * 100
    porcentaje_sabado = (NS_ventas / total_ventas) * 100
else:
    porcentaje_lunes = 0
    porcentaje_martes = 0
    porcentaje_miercoles = 0
    porcentaje_sabado = 0

print("Total vendido en la semana: ",total_semana)
print("Promedio vendido el miércoles: ",promedio_miercoles)
print("Promedio vendido en la semana: ",promedio_semana)
print("Porcentaje de ventas el lunes: %",porcentaje_lunes)
print("Porcentaje de ventas el martes: %",porcentaje_martes)
print("Porcentaje de ventas el miércoles: %",porcentaje_miercoles)
print("Porcentaje de ventas el sábado: %",porcentaje_sabado)
```

●	Para las operaciones necesarias puede declarar las variables que desee.

El siguiente pseudocódigo calcula la fuerza gravitatoria (fórmula establecida y descrita por Isaac Newton), ¿Cuáles son los resultados de las variables m1, m2, d1 y F? ¿Qué tipo de dato contiene F?

m1 = (45 + 3) % 2

m2 = 2 + 3 * 4 * 0.5

d1 = (3 % 2) - (3 / 2)

G = 9.8

image.png



```python
m1 = (45 + 3) % 2
m2 = 2 + 3 * 4 * 0.5
d1 = (3 % 2) - (3 / 2)
G = 9.8

print("m1 =", m1)
print("m2 =", m2)
print("d1 =", d1)
print("G =", G)

F = (G * m1 * m2) / (d1 * d1)

print("La fuerza gravitacional es:", F)
print("El tipo de dato de F es:", type(F))
```
