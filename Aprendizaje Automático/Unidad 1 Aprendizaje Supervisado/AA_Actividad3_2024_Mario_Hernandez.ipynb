## Problema 3

Implemente el algoritmo de _Gradient Descent_ para obtener los parámetros de un modelo lineal para la regresión de la etiqueta $y$ a partir de las características $x$.

a) Implemente una función _predict_ que calcule el valor de $\hat{y}$ dado el valor de las características $x$ y los parámetros $b$ y $w$ según el modelo:
$$\hat{y} = b + w\cdot x$$

```python
# Implemente la función predict, de tal manera que pueda realizar varias predicciones a la vez.
# En particular, dada una matriz X, la función debe retornar tantas predicciones como filas tiene X.
# Asegúrese de validar las dimensiones de X y w. En caso de que las dimensiones no correspondan,
# su función debe arrojar un RuntimeError

def predict(X, b, w):
    if X.shape[1] != len(w):
        raise RuntimeError(f"Dimensiones incompatibles: X shape {X.shape}, w length {len(w)}")
    return b + np.dot(X, w)
```

```python
# Probar la función realizando una predicción para los datos de prueba utilizando los
# parámetros b = 1 y w = 1 (para cada componente)

# **** No cambiar celda, sólo ejecutar. ****

X_val = np.array([[ 1. ,  0. , 95.3, 58.2, 89.5, 36. , 71.5, 52.],
       [ 5. ,  1. , 90.4, 55.8, 86. , 36.5, 63.2, 44.2]])

yp_val = predict(X_val, 1, np.ones(X_val.shape[1]))

expected = np.array([404.5, 383.1])

diff = np.sum(expected - yp_val)
if diff == 0:
  print('La función devolvió los valores esperados')
else:
  print('La función NO devolvió los valores esperados')
```

b) Implemente una función para medir el _Mean Squared Error_ (MSE) de un par de vectores $y \in \mathbb{R}^n$ y $\hat{y} \in \mathbb{R}^n$.

```python
def mse(y, yp):
    return np.mean((y - yp) ** 2)
```

c) Implemente una función para calcular la derivada del MSE con respecto a $b$ y $w$. La función debe devolver una tupla $(gb, gw)$ dónde $gb$ es un escalar y $gw$ es un vector de la misma dimensión que $w$.

```python
def gradient(X, y, b, w):
    yp = predict(X, b, w)
    error = yp - y
    gb = np.mean(error)  # derivada respecto a b
    gw = np.mean(X * error[:, np.newaxis], axis=0)  # derivada respecto a w
    return gb, gw
```

d) Implemente el algoritmo de _Gradient Descent_ de tal manera que dadas las características $X \in \mathbb{R}^{n \times m}$ y las etiquetas $y \in \mathbb{R}^n$, devuelva los parámetros $b \in \mathbb{R}$ y $w \in \mathbb{R}^m$, que representan los parámetros de un modelo de regresión lineal que minimiza el MSE del modelo con respecto a los datos dados.

Adicionalmente, incluya los parámetros:

*   $alpha \in \mathbb{R}$: velocidad de aprendizaje
*   $eps$: si la diferencia entre los errores de dos iteraciones contiguas es menor a este valor, el proceso se detiene aún si quedan iteraciones. El valor por defecto es None que indica que no aplica este criterio.
*   $maxIter$: número máximo de iteraciones a realizar



```python
def gradient_descent(X, y, alpha=0.00001, eps=None, maxIter=100):
    # Inicializar parámetros -> algoritmo de Gradient Descent
    b = 0
    w = np.zeros(X.shape[1])

    # Calcular error inicial
    err = mse(y, predict(X, b, w))
    print("Iteration: ", 0, " error:", err)

    # Iterar hasta convergencia o máximo de iteraciones
    prev_err = float('inf')
    for i in range(maxIter):
        # Calcular gradientes
        gb, gw = gradient(X, y, b, w)

        # Actualizar parámetros
        b = b - alpha * gb
        w = w - alpha * gw

        # Calcular nuevo error
        err = mse(y, predict(X, b, w))
        print("Iteration:", i+1, "error:", err)

        # Verificar convergencia si eps está definido
        if eps is not None and abs(prev_err - err) < eps:
            print("Convergencia alcanzada en iteración", i+1)
            break

        prev_err = err

    return b, w
```

```python
# Entrene el modelo utilizando X_train y y_train. Guarde los parámetros resultantes
b, w = gradient_descent(X_train, y_train, alpha=0.00001, maxIter=1000)
```

```python
# Reporte el MSE en que incurre el modelo con los datos de entrenamiento
y_train_pred = predict(X_train, b, w)
mse_train = mse(y_train, y_train_pred)
print("\nMSE en conjunto de entrenamiento:", mse_train)
```

```python
# Importamos las librerías necesarias
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Creamos y entrenamos el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Realizamos las predicciones en los datos de entrenamiento
y_train_pred_linear = model.predict(X_train)
mse_train_linear = mean_squared_error(y_train, y_train_pred_linear)
print("\nMSE en conjunto de entrenamiento (LinearRegression):", mse_train_linear)

# Realizamos las predicciones en los datos de prueba
y_test_pred_linear = model.predict(X_test)
mse_test_linear = mean_squared_error(y_test, y_test_pred_linear)
print("MSE en conjunto de prueba (LinearRegression):", mse_test_linear)

zari_age_linear = model.predict(Zari_features)[0]
print('\nEdad estimada de Zari con LinearRegression:', zari_age_linear)
```

```python
# Importamos el modelo de Árbol de Decisión y la métrica MSE
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Creamos el modelo de Árbol de Decisión
tree_model = DecisionTreeRegressor(random_state=41, max_depth=5)  # limitamos la profundidad para evitar overfitting

# Entrenamos el modelo
tree_model.fit(X_train, y_train)

# Obtenemos las predicciones para el conjunto de entrenamiento
y_train_pred_tree = tree_model.predict(X_train)
mse_train_tree = mean_squared_error(y_train, y_train_pred_tree)
print("\nMSE en conjunto de entrenamiento (Árbol de Decisión):", mse_train_tree)

# Obtenemos las predicciones para el conjunto de prueba
y_test_pred_tree = tree_model.predict(X_test)
mse_test_tree = mean_squared_error(y_test, y_test_pred_tree)
print("MSE en conjunto de prueba (Árbol de Decisión):", mse_test_tree)

zari_age_tree = tree_model.predict(Zari_features)[0]
print('\nEdad estimada de Zari con Árbol de Decisión:', zari_age_tree)
```
