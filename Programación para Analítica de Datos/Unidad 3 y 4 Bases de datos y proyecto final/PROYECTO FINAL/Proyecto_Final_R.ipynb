Este conjunto de datos contiene una lista exhaustiva de las canciones m치s famosas de 2023 que aparecen en Spotify. El conjunto de datos ofrece una gran cantidad de caracter칤sticas m치s all치 de lo que suele estar disponible en conjuntos de datos similares. Proporciona informaci칩n sobre los atributos de cada canci칩n, su popularidad y su presencia en diversas plataformas musicales. El conjunto de datos incluye informaci칩n como el nombre de la canci칩n, el nombre del artista o artistas, la fecha de lanzamiento, listas de reproducci칩n y listas de Spotify, estad칤sticas de streaming, presencia en Apple Music, presencia en Deezer, listas de Shazam y varias caracter칤sticas de audio.

[Fuente](https://www.kaggle.com/datasets/nelgiriyewithana/top-spotify-songs-2023/datahttps://)

* track_name: Nombre de la canci칩n
* artist(s)_name: Nombre del artista o artistas
* artist_count: N칰mero de artistas
* released_year: A침o de lanzamiento
* released_month: Mes de lanzamiento
* released_day: D칤a de lanzamiento
* in_spotify_playlists: N칰mero de playlists de Spotify en las que aparece la canci칩n
* in_spotify_charts: N칰mero de veces que la canci칩n ha aparecido en los charts de Spotify
* streams: N칰mero de streams
* in_apple_playlists: N칰mero de playlists de Apple en las que aparece la canci칩n
* in_apple_charts: N칰mero de veces que la canci칩n ha aparecido en los charts de Apple
* in_deezer_playlists: N칰mero de playlists de Deezer en las que aparece la canci칩n
* in_deezer_charts: N칰mero de veces que la canci칩n ha aparecido en los charts de Deezer
* bpm: Beats per minute
* key: Tonalidad
* mode: Modo (mayor o menor)
* danceability_%: Porcentaje de bailabilidad
* valence_%: Porcentaje de positividad
* energy_%: Porcentaje de energ칤a
* acousticness_%: Porcentaje de ac칰stica
* instrumentalness_%: Porcentaje de instrumentalidad
* liveness_%: Porcentaje de viveza
* speechiness_%: Porcentaje de discurso

```r
# Definimos las variables para PCA
X <- df[, c("danceability_", "energy_", "acousticness_", "valence_", "bpm")]

# Escalamos los datos
X_scaled <- scale(X)

# PCA
pca <- prcomp(X_scaled, center = TRUE, scale. = TRUE)

# Obtenemos las componentes principales y la varianza explicada
componentes_principales <- pca$x[, 1:2]  # Tomamos las primeras 2 componentes principales
varianza_explicada <- pca$sdev^2 / sum(pca$sdev^2)

print("Varianza explicada por cada componente:")
print(varianza_explicada)

print("Componentes principales:")
print(pca$rotation[, 1:2])
```

```r
# Seleccionamos solo las variables num칠ricas para el PCA
numeric_cols <- c('danceability_', 'valence_', 'energy_',
                  'acousticness_', 'bpm')

X <- df[, numeric_cols]

# Escalamos los datos
scaled_data <- scale(X)

# PCA
pca <- prcomp(scaled_data, center = TRUE, scale. = TRUE)

# Obtenemos las componentes principales
principal_components <- pca$x

# Varianza explicada por cada componente
explained_variance_ratio <- pca$sdev^2 / sum(pca$sdev^2)

# Graficamos la varianza explicada acumulada
num_components <- length(explained_variance_ratio)
cumulative_variance <- cumsum(explained_variance_ratio)

ggplot(data = data.frame(components = 1:num_components, cumulative_variance = cumulative_variance),
       aes(x = components, y = cumulative_variance)) +
  geom_line(color = "green", linetype = "dashed") +  # Geometr칤a de l칤nea
  geom_point(shape = 16, color = "green") +          # Geometr칤a de punto con shape 16 (c칤rculo s칩lido)
  labs(title = "Varianza Explicada Acumulada por Componentes Principales",
       x = "N칰mero de Componentes Principales",
       y = "Varianza Explicada Acumulada") +
  scale_x_continuous(breaks = seq(1, num_components, by = 1)) +
  theme_minimal()

print(ggplot)
```
