pandas es un paquete de Python que proporciona estructuras de datos similares a los dataframes de R. Pandas depende de Numpy, la librería que añade un potente tipo matricial a Python. Los principales tipos de datos que pueden representarse con pandas son:

    -Datos tabulares con columnas de tipo heterogéneo con etiquetas en columnas y filas.

    -Series temporales.

Pandas proporciona herramientas que permiten:

    -leer y escribir datos en diferentes formatos: CSV, Microsoft Excel, bases SQL y formato HDF5

    -seleccionar y filtrar de manera sencilla tablas de datos en función de posición, valor o etiquetas

    -fusionar y unir datos

    -transformar datos aplicando funciones tanto en global como por ventanas

    -manipulación de series temporales

    -hacer gráficas

En pandas existen tres tipos básicos de objetos todos ellos basados a su vez en Numpy:

Series (listas, 1D),
DataFrame (tablas, 2D) y![archivo.png]
Panels (tablas 3D).

Para facilitar el trabajo, algunas veces podemos reorganizar algunos formatos de columnas en un dataframe, por ejemplo, la siguiente linea nos permita crear una columna con un formato diferente de fecha

```python
d.index = pd.to_datetime(d.Date, dayfirst=True)
d.head()
```

También podemos eliminar completamente columnas redudantes o que nos aporten poca información, por ejemplo las columnas "Date" o "Unnamed: 1"

```python
del(d["Date"])
del(d["Unnamed: 1"])
d.head()
```

Con la reorganización de la fecha los indices quedaron definidos por fechas y no solo por números como se muestra continuación:

```python
d.index
```

Podemos seguir eligiendo las columnas que nos interesen y cambiarles el nombre, como se muestra en la siguiente línea

```python
d.columns=["Berri", "Mneuve1", "Mneuve2", "Brebeuf", "Ecuacion Berri"]
d.head()
```

La función sum() nos permite contar la cantidad de elementos que hay con una característica en una columna, en el siguiente ejemplo se muestra como se cuentan los datos no validos en cada columna, utilizando **col** como variable de iteración que recorre cada columna del dataframe

```python
for col in d.columns:
    print(col, np.sum(pd.isnull(d[col])))
```

Cuando encontremos valores no validos en una columna, podemos utilizar el método fillna() para reemplizarlos por un valor o cadena, en el siguiente ejemplo se reemplazó por el promedio de los datos validos de la columna Brebeuf

```python
d.Brebeuf.fillna(d.Brebeuf.mean(), inplace=True)
```

Despues del reemplazo podemos verificar como cambió el datframe

```python
d.head(365)
```

El método sort nos permite reorganizar los datos de un dataframe en orden ascendente, en este caso se hizo siguiendo las fechas del indice

## Filtrado de un dataframe

## Construcción de arreglos y listas

Las matrices pueden convertirse en dataframe, definiendo el nombre de las columnas y el número de indice con una lista de los números de 10 hasta el tamaño de la matriz inicial.
